```python
# Функция загрузки Данных с яндекс Диска

API_ENDPOINT = 'https://cloud-api.yandex.net/v1/disk/public/resources/download?public_key={}'


def _get_real_direct_link(sharing_link):
    pk_request = requests.get(API_ENDPOINT.format(sharing_link))
    
    # Returns None if the link cannot be "converted"
    return pk_request.json().get('href')


def _extract_filename(direct_link):
    for chunk in direct_link.strip().split('&'):
        if chunk.startswith('filename='):
            return chunk.split('=')[1]
    return None


def download_yadisk_link(sharing_link, filename=None):
    direct_link = _get_real_direct_link(sharing_link)
    if direct_link:
        # Try to recover the filename from the link
        filename = filename or _extract_filename(direct_link)
        
        download = requests.get(direct_link)
        with open(filename, 'wb') as out_file:
            out_file.write(download.content)
        print('Downloaded "{}" to "{}"'.format(sharing_link, filename))
    else:
        print('Failed to download "{}"'.format(sharing_link))
```

Для того чтобы найти код проекта воспользуемся шаблонами, по которым будем искать необходимые значения  
Используем библиотеку <div style="font-size: 22px"> **re**

```python
# Шаблоны для поиска Кода проекта

# Полный формат
pattern_1 = r'[A-Za-z]{3}[-]{1}[A-Za-z]{3}[-]{1}[A-Za-z]{1}[-]{1}[0-9]{5}[-]{1}[0-9]{2}[-]{1}[A-Za-z]{1}'
# Минимальный формат
pattern_2 = r'[A-Za-z]{3}[-]{1}[A-Za-z]{3}[-]{1}[A-Za-z]{1}[-]{1}[0-9]{5}'


# Шаблон формата короче мин.допустимого но они также встречаются
pattern_3 = r'[A-Z]{2}[-]{1}[A-Z]{3}[-]{1}[0-9]{3}'
pattern_4 = r'[A-Z]{2}[-]{1}[0-9]{5}'


# Поместим все шаблоны в список
list_patterns = [pattern_1, pattern_2, pattern_3, pattern_4]
```

```python
# Функция для поиска кода проекта

def project_code(row):
    string = row['Project'] + row['Task'] + row['Description'] + row['Parent_Task']
    
    for n, pat in enumerate(list_patterns):
        try:
            if pat in list_patterns:
                return re.search(pat, string)[0].upper()
        except: continue
```

```python
# Новый атрибут 'Project_code' 

df_0['Project_code'] = df_0.apply(project_code, axis=1)
```

Чтобы указать того же заказчика, который не указан в остальных задачах в рамках одного проекта, необходимо найти все задачи, которые составляют один портфель.  
Минимальный формат 'Project_code' укзазывает на проект, в состав которого входит несколько подзадач

```python
# Шаблоны минимального формата

pattern_5 = r'[A-Z]{3}[-]{1}[A-Z]{3}[-]{1}[A-Z]{1}[-]{1}[0-9]{5}'
pattern_6 = r'[A-Z]{2}[-]{1}[A-Z]{3}'
pattern_7 = r'[A-Z]{2}[-]{1}[0-9]{5}'
list_patterns_2 = [pattern_5, pattern_6, pattern_7]
```

```python
# Функция поиска минимального формата кода проекта

def min_code(row):
    string = row['Project_code']
    
    for n, pat in enumerate(list_patterns_2):
        try:
            if pat in list_patterns_2:
                return re.search(pat, string)[0]
        except: continue
```

```python
df_0['min_code'] = df_0.apply(min_code, axis=1)
```

```python
# Используем минимальный формат кода как метку, по которой будем находить заказчика для подзадач

for code in df_0['min_code'].unique():
    requester = (df_0[(df_0['min_code'] == code)&df_0['Requester'].notna()]['Requester']).unique()
    
    try:
        df_0.loc[(df_0['min_code'] == code)&(df_0['Requester'].isna()), 'Requester'] = requester[0]
    except:
        continue
```

```python
# Функция для подсчета часов по значению минут (Общее списанное время также отображается в Decimal_Hours)

def hours_minutes(row):
    
    # Количество целых часов
    row['Hours'] = row['Hours'] + row['Minutes']//60
    # Остаток минут
    row['Minutes'] = row['Minutes']%60
    return row

workload_pivot = workload_pivot.apply(hours_minutes, axis=1)
```

```python
# Добавим количество проектов для каждого исполнителя

workload_pivot['Number_of_projects'] = df_0.pivot_table(index = 'Who', values = 'Id', aggfunc='count')
```

```python
# Добавим статус проекта ('Is_it_Billable?') - Олачиваемый

workload_pivot['Billable'] = df_0.pivot_table(index='Who', values='Is_it_Billable?', aggfunc='sum')
```

```python
# Добавим статус проекта ('Is_it_Billable?') - НеОлачиваемый

workload_pivot['Not_Billable'] = workload_pivot['Number_of_projects'] - workload_pivot['Billable']
```

```python
# Добавим соотношение Billable и Not_Billable

workload_pivot['Billable_rate'] = workload_pivot['Not_Billable']/workload_pivot['Billable']
```

### Время выпуска отчетов

#### Выпуск отчета в последний день отчетного периода

```python
# Определение Дня недели по дате

def weekday(row):
    return row['Date'].dayofweek
df_0['Weekday'] = df_0.apply(weekday, axis=1)
```

```python
# Функция наименования дней недели

list_weekday = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

def naming_weekdays (row):

    for n, day in enumerate(list_weekday):
        if n == row['Weekday']:
            return list_weekday[n]
        else:
            pass

df_0['Day_of_week'] = df_0.apply(naming_weekdays, axis=1)
```

```python
# Шаблон для 20 числа каждого месяца

pattern_8 = r'[0-9]{4}[-]{1}[0-9]{2}[-]{1}20'
pattern_9 = r'[0-9]{4}[-]{1}[0-9]{2}[-]{1}(?!20$)([0-9]{2})'
```

```python
list_patterns_3 = [pattern_8, pattern_9]
```

```python
# Функция для поиска выгрузки отчета 20 числа каждого месяца

def last_day_report(row):
    string = str(row['Date'])
    for n, pat in enumerate(list_patterns_3):
        try:
            if pat == pattern_8 and re.search(pat, string)[0] in string:
                last_day = 1
            else:
                last_day = 0
        except:
            continue
        return last_day
```

```python
df_0['Last_day_report'] = df_0.apply(last_day_report, axis=1)
```

```python
# Выпуск отчета в последний момент.

df_0[df_0['Last_day_report'] == 1]['Who'].value_counts()
```

```python
# Добавим в сводную таблицу

workload_pivot['Last_day_report'] = df_0.pivot_table(index='Who', values='Last_day_report', aggfunc='sum')
```

```python
# Зависимость выпуска отчета в последний день от других признаков

phik_cols = ['Hours', 'Number_of_projects', 'Billable', 'Not_Billable', 'Last_day_report']

barplot(round(workload_pivot[phik_cols].phik_matrix()['Last_day_report'].sort_values(ascending=False), 3),
       suptitle = 'Влияние Факторов на выпуск отчета в последний день',
        xlabel='Факторы',
        ylabel='Корреляция'
       )
plt.xticks(rotation=75)
plt.show()
```
